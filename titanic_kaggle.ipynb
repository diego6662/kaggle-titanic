```python
import pandas as pd
import numpy as np
```

```python
# import the train set
train = pd.read_csv('train.csv')
train.head(15)
```

```python
train.drop(['Name','PassengerId','Ticket'],axis=1, inplace=True)
#for the moment i keep the feature Embarked, i will analyze the importance of this feature
train.head()
```

```python
y_train = train['Survived']
X_train = train.drop('Survived',axis=1)
# Separed the label from the dataset
```

```python
X_train.head(10)
```

## I start the data preproccesing

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
encoder = OneHotEncoder(categories=[0])
label = LabelEncoder()
X_train['Sex'] = label.fit_transform(X_train['Sex'].to_numpy()) #I change the values for male = 1 and female = 0
X_train.head()
```

```python
X_train['Cabin'].isnull().sum()
```

```python
most_repeated = X_train['Cabin'].value_counts()[:1].index.tolist()
X_train['Cabin'].fillna(most_repeated[0],inplace=True)
X_train['Cabin'].isnull().sum()
```

```python
X_train['Cabin'] = label.fit_transform(X_train['Cabin'].to_numpy())
X_train.head()
```

```python
X_train['Embarked'].isnull().sum()
```

```python
most_repeated = X_train['Embarked'].value_counts()[:1].index.tolist()
X_train['Embarked'].fillna(most_repeated[0],inplace=True)
X_train['Embarked'] = label.fit_transform(X_train['Embarked'].to_numpy())
X_train['Embarked'].isnull().sum()
```

```python
X_train.head()
```

```python
print(X_train['Fare'].isnull().sum())
print(X_train['Parch'].isnull().sum())
print(X_train['SibSp'].isnull().sum())
print(X_train['Age'].isnull().sum())
print(X_train['Sex'].isnull().sum())
print(X_train['Pclass'].isnull().sum())
```

```python
median = X_train['Age'].median()
X_train['Age'].fillna(median,inplace=True)
print(X_train['Age'].isnull().sum())
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train['Pclass'] = scaler.fit_transform(X_train['Pclass'].to_numpy().reshape(-1,1))
X_train['Age'] = scaler.fit_transform(X_train['Age'].to_numpy().reshape(-1,1))
X_train['SibSp'] = scaler.fit_transform(X_train['SibSp'].to_numpy().reshape(-1,1))
X_train['Parch'] = scaler.fit_transform(X_train['Parch'].to_numpy().reshape(-1,1))
X_train['Fare'] = scaler.fit_transform(X_train['Fare'].to_numpy().reshape(-1,1))
X_train['Cabin'] = scaler.fit_transform(X_train['Cabin'].to_numpy().reshape(-1,1))
X_train['Embarked'] = scaler.fit_transform(X_train['Embarked'].to_numpy().reshape(-1,1))
X_train.head()
```
